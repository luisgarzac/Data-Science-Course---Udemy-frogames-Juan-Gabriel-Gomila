# Carga de datos a través de la función open

```python
data3 = open(mainpath + "/" + "customer-churn-model/Customer Churn Model.txt", "r") # "r" es para modo lectura, "w" es para modo escritura, "a" es modo append para agregar contenido y no borrara contenido
```

```python
cols = data3.readline().strip().split(",") # leemos cada linea, eliminamos espacios en blanco y agregamos una coma entre elementos
n_cols = len(cols)
```

```python
counter = 0 #creamos un contador
main_dict = {} #creamos un diccionario
for col in cols:
    main_dict[col] = []
```

```python
for line in data3:
    values = line.strip().split(",") # quitamos espacios en blanco y dividimos por comas
    for i in range(len(cols)): #itera sobre columnas de cada linea y añade el nombre de la columna respectiva del diccionario
        main_dict[cols[i]].append(values[i]) 
    counter +=1 # cada que lee una fila 
print("dataset tiene %d filas y %d columnas" % (counter, n_cols))
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

# Lectura y Escritura de ficheros

```python
infile = mainpath + "/" + "customer-churn-model/Customer Churn Model.txt"
outfile = mainpath + "/" + "customer-churn-model/Tab Customer Churn Model.txt"
```

```python
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1: 
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```

```python
df4 = pd.read_csv(outfile, sep = "\t")
df4.head()
```
